```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from pathlib import Path
import os
import json
from bddutils import DataHandler
```

```python
my_handler = DataHandler('./data/bdd100k/')
```

```python
my_handler.get_random(5)
```

```python
# paths 
images_train = Path('./data/bdd100k/images/100k/train')
labels_train = Path('./data/bdd100k/labels/bdd100k_labels_images_train.json')
```

```python
# load in the labels
with open(labels_train, 'rb') as f:
    labels = json.load(f)
```

```python
labels[0]['name']
```

```python
# load in image data
img_locs = os.listdir(images_train)
```

```python
# im to read
start = 17000
add = 5
for i in range(start, start+add):
    img_num = i

    t1 = plt.imread(images_train / labels[img_num]['name'])
    print(t1.shape)
    print(img_locs[2000])
    fig, ax = plt.subplots(1, figsize=(12,10))
    ax.imshow(t1)
    # Create a Rectangle patch
    rects = []
    for objects in labels[img_num]['labels']:
        try:
            x1, y1, x2, y2 = objects['box2d']['x1'], objects['box2d']['y1'], objects['box2d']['x2'], objects['box2d']['y2'] 
        except:
            continue
        ax.add_patch(patches.Rectangle((x1, y1), x2-x1, y2-y1, linewidth=1, edgecolor='r', facecolor='None'))
        ax.annotate(objects['category'], color='w', xy=(x1, y1), fontsize=8)
    plt.xticks([])
    plt.yticks([])
```
