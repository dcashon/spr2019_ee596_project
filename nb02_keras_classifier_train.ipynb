```python
from kmodel import build_model_accurate
from kutils import batcher
from pathlib import Path
import numpy as np
from keras.utils import to_categorical
import os
import keras.optimizers
import numpy as np
from sklearn.utils import shuffle
```

```python
# get model
model = build_model_accurate()
data_path = Path('/mnt/disks/gscratch/train')
```

```python
#weighting for imbalanced data
w = [2.99525862e+01, 7.21224702e+00, 1.48799263e-01, 4.61727575e+01,
       1.07652982e+00, 8.08023256e+01, 3.27551261e+00, 6.84967965e-01,
       1.98542857e+01, 2.63969611e+00]
y = np.arange(10)
weights = dict(zip(y,w))
```

```python
model.summary()
```

```python
# ------ MODEL TRAINING -------
# configuration
opt = keras.optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
opt2 = keras.optimizers.rmsprop(lr=0.0001, decay=1e-6)

model.compile(optimizer=opt2,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# TRAIN
num_epochs = 20
num_data = 13
batch_size = 32
counter = 0
for i in range(num_epochs):
    print('Training on Epoch \t %s' % i)
    for j in range(num_data):
        print('Training on DataBatch Num %s' % j)
        c_batcher = batcher(j, batch_size, data_path)
        print ('Loss \t Acc')
        for data, labels in c_batcher:
            counter += 1
            labels_reshape = to_categorical(labels, num_classes=10)
            data_shuffle, labels_shuffle = shuffle(data, labels_reshape)
            t = model.train_on_batch(data_shuffle.astype('float32') / 255, labels_shuffle.astype('float32'), class_weight=weights)
            if counter % 10 == 0:
                print(str(t[0]) + '\t' + str(t[1]))
            if counter % 100 == 0:
                w_test = model.get_weights()
                print(w_test[0][0, :, 0, 0])
    model.save('/mnt/disks/gscratch/models/classifier3_epoch' + str(i) + '.h5')
```
