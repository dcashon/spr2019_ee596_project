```python
import numpy as np
import matplotlib.pyplot as plt
from kmodel import build_model_classifier, build_model_regressor
import keras
from pathlib import Path
import os
from kutils import batcher
from keras import backend as K
from keras import losses
```

```python
# create the regressor model by appending layers to the classification feature map
# load latest model
classifier_path = '/mnt/disks/gscratch/models/my_model4_epoch4.h5'
regressor = build_model_regressor(classifier_path)
```

```python
regressor.summary()
```

```python
# model params
data_path = Path('/mnt/disks/gscratch/train')
opt = keras.optimizers.rmsprop(lr=0.001, decay=1e-6)
regressor.compile(optimizer=opt,
              loss=losses.mean_squared_error)
```

```python
# -----------Train
num_epochs = 25
num_data = 13
batch_size = 32
counter = 0
for i in range(num_epochs):
    print('Training on Epoch \t %s' % i)
    for j in range(num_data):
        print('Training on DataBatch Num %s' % j)
        c_batcher = batcher(j, batch_size, data_path, classify=False)
        print ('Loss')
        for data, labels in c_batcher:
            counter += 1
            t = regressor.train_on_batch(data / 255, np.reshape(labels, (-1, 1, 1, 4)))
            #w_test = regressor.get_weights()
            if counter % 10 == 0:
                print(t)
    regressor.save('/mnt/disks/gscratch/models/regressor4_epoch' + str(i) + '.h5')
```
