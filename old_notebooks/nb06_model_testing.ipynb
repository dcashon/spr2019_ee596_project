```python
from bddmodel import BDDModel, SimpleBDDModel
import tensorflow as tf
from bddfunc import training_batcher
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import datetime
```

```python
# Initialize model
my_model = SimpleBDDModel()
```

```python
first_train_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES,
                                     "classifier")
```

```python
# hyperparameters
batch_size = 128
num_data = 13 # number of saved batch data
num_epochs = 5
```

```python
# Train Model
init = tf.global_variables_initializer()
saver = tf.train.Saver()
enc = OneHotEncoder(sparse=False, categories=[np.arange(10)])
day_path = str(datetime.datetime.now())[:10]
counter = 0
with tf.Session() as sess:
    init.run()
    for i in range(num_epochs):
        print('Training on Epoch \t' + str(i))
        for j in range(13): # 13 saved data
            print('Training on Dataset \t' + str(j))
            print('Loss \t Acc')
            batcher = training_batcher(j, batch_size)
            for data, labels in batcher:
                counter += 1
                temp = enc.fit_transform(np.reshape(labels, (-1,1)))
                f_dict = {my_model.X: data / 255, my_model.Y1: temp} #my_model.training:True}
                # update the operations to accomadate batch-norm
                #extra_update_ops = tf.get_collection(tf.GraphKeys.UPDATE_OPS)
                #train_op = my_model.minimizer
                #sess.run([train_op, extra_update_ops], feed_dict=f_dict)
                my_model.minimizer.run(feed_dict=f_dict)
                if counter % 10 == 0:
                    print(str(my_model.loss.eval(feed_dict=f_dict)) + '\t' + str(my_model.accuracy.eval(feed_dict=f_dict)) )
        save_path = saver.save(sess, "./tmp/model3_" + day_path + "_epoch" + str(i) + ".ckpt")
        print("Model saved in path: %s" % save_path)
```
